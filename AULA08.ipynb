```python
import random
def jogar():
    print("Escolha o nível de dificuldade:")
    print("(1) Fácil [20 tentativas] (2) Médio [15 tentativas] (3) Difícil [10 tentativas]")
    
    nivel = int(input("Define o nível: "))
    
    if nivel == 1:
        total_tentativas = 20
    elif nivel == 2:
        total_tentativas = 15
    else:
        total_tentativas = 10

    numero_secreto = random.randint(1, 50)
    
    for rodada in range(1, total_tentativas + 1):
        print(f"\nTentativa {rodada} de {total_tentativas}")
        chute = int(input("Digite um número entre 1 e 50: "))

        if chute < 1 or chute > 50:
            print("Digite um numero entre 1 e 50!")
            continue

        acertou = chute == numero_secreto
        maior   = chute > numero_secreto
        menor   = chute < numero_secreto

        if acertou:
            print(f"Você acertou! O número era {numero_secreto}.")
            break
        else:
            if maior:
                print("Você errou! O seu chute foi maior que o número secreto.")
            elif menor:
                print("Você errou! O seu chute foi menor que o número secreto.")
        
        if rodada == total_tentativas:
            print(f"\nGame Over! O número era {numero_secreto}.")

jogar()
```
